```python
#multiple class function 
class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []   # list of dictionaries with 'title' and 'artist'

    def add_song(self, title, artist):
        self.songs.append({"title": title, "artist": artist})
        print(f"Added: {title} by {artist}")

    def remove_song(self, title):
        for song in self.songs:
            if song["title"] == title:
                self.songs.remove(song)
                print(f"Removed: {title}")
                return
        print(f"{title} not found in playlist.")

    def show_songs(self):
        print(f"\nPlaylist '{self.name}':")
        for song in self.songs:
            print(f"- {song['title']} (Artist: {song['artist']})")

    def add_multiple_songs(self):
        n = int(input("How many songs do you want to add? "))
        for i in range(n):
            title = input(f"add the song {i+1}: ")
            artist = input(f"Enter artist for song {i+1}: ")
            self.add_song(title, artist)


# Example usage
my_playlist = Playlist("Favorites")
my_playlist.add_multiple_songs()
my_playlist.show_songs()
```
